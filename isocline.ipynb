```python
import numpy as np
import matplotlib.pyplot as mp
from jupyterthemes import jtplot
```

```python
jtplot.style()
mp.rcParams["text.usetex"] = True
mp.rcParams["font.family"] = "serif"
mp.rcParams["figure.figsize"] = (5, 5)
mp.rcParams["figure.dpi"] = 100
```

```python
x = np.linspace(-2,2,500)

mp.plot(x, (1-x**2), "w-")
mp.plot(x, -(1-x**2), "w-")

mp.plot(x, 2*(1-x**2), "w-")
mp.plot(x, -2*(1-x**2), "w-")

mp.plot(x, 1/2*(1-x**2), "w-")
mp.plot(x, -1/2*(1-x**2), "w-")

mp.plot(x*0+1, x, "w-")
mp.plot(x*0-1, x, "w-")

mp.plot(x, x*0, "w-")

mp.xlim(-2, 2)
mp.ylim(-2, 2)

E = lambda y, ddy: ddy -1 + y**2

X,Y = np.meshgrid(x,x)
Z = E(X,Y)

mp.contourf(X,Y,Z, cmap="OrRd_r")
mp.colorbar()
mp.show()

mp.show()
```

```python
x = np.linspace(-2,2,500)

mp.plot(x, -abs(x), "w-")

mp.plot(x, -1/2*abs(x), "w-")

mp.plot(x, x*0, "w-")
mp.plot(x*0, x, "w-")

mp.plot(x, -2*abs(x), "w-")

mp.xlim(-2, 2)
mp.ylim(-2, 2)

E = lambda th, dth: 2*dth + abs(th)

X,Y = np.meshgrid(x,x)
Z = E(X,Y)

mp.contourf(X,Y,Z, cmap="OrRd_r")
mp.colorbar()

mp.show()
```
