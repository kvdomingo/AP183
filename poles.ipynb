```python
import numpy as np
import matplotlib.pyplot as mp
from jupyterthemes import jtplot
```

```python
jtplot.reset()
mp.rcParams["font.family"] = "serif"
mp.rcParams["text.usetex"] = True
mp.rcParams["figure.figsize"] = (5*16/9, 5)
mp.rcParams["figure.dpi"] = 100
```

```python
class polesystem:
    
    def __init__(self, t, wn, zeta):
        self.theta = np.arctan2(zeta, np.sqrt(1 - zeta**2))
        self.Real = -zeta*wn
        self.Imag = wn*np.sqrt(1 - zeta**2)
        self.oldSys = 1 - np.exp(self.Real*t)/np.sqrt(1 - zeta**2) * np.cos(self.Imag*t - self.theta)
        self.t = t
        self.wn = wn
        self.zeta = zeta
        self.newSys = None
        
    def moveRealPole(self, sigma):
        self.Real += sigma
        self.updateSys()
        
    def moveImagPole(self, jw):
        self.Imag += jw
        self.updateSys()
        
    def updateSys(self):
        self.sys = 1 - np.exp(self.Real*t)/np.sqrt(1 - zeta**2) * np.cos(self.Imag*t - self.theta)
```

```python
t = np.linspace(0, 10*np.pi, 1000)
wn = 10
zeta = 0.01
c = polesystem(t, wn, zeta)
mp.plot(t, c.oldSys, "r-", alpha=0.6)

c.moveRealPole(+0.05)
mp.plot(t, c.sys, "k-")

mp.savefig("original.png", dpi=300, bbox_inches="tight")
mp.show()
```

```python
c = polesystem(t, wn, zeta)
c.moveRealPole(+0.05)
mp.plot(t, c.sys, "b-")
#mp.savefig("minusreal.png", dpi=300, bbox_inches="tight")
mp.show()
```

```python
c = polesystem(t, wn, zeta)
c.moveImagPole(10)
mp.plot(t, c.sys, "b-")
mp.savefig("plusimag.png", dpi=300, bbox_inches="tight")
mp.show()
```

```python
c = polesystem(t, wn, zeta)
c.moveImagPole(10)
c.moveRealPole(-0.1)
mp.plot(t, c.sys, "b-")
mp.savefig("plusangle.png", dpi=300, bbox_inches="tight")
mp.show()
```

```python
c = polesystem(t, wn, zeta)
c.moveRealPole(0.05)
mp.plot(t, c.sys, "b-")
mp.savefig("plusreal.png", dpi=300, bbox_inches="tight")
mp.show()
```
