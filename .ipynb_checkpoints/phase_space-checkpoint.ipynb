```python
import numpy as np
import matplotlib.pyplot as mp
from jupyterthemes import jtplot
```

```python
jtplot.reset()
mp.rcParams["font.family"] = "serif"
mp.rcParams["text.usetex"] = True
mp.rcParams["figure.figsize"] = (5, 5)
mp.rcParams["figure.dpi"] = 100
```

```python
class rksolve:
    
    def __init__(self,f):
        self.f = f
        self.init_conds = None
        self.phase_solution = None
    
    def rk4(self,ta,tb,N=1000):
        """
        Perform Runge-Kutta stage-4 (RK4) integration of a time-varying system for
        an arbitrary number of variables/dimensions. Use rk4_spec to specify exact
        step size.

        Parameters
        ----------
        ta : float
            Time to begin integration
        tb : float
            Time to end integration
        N : int
            Number of integration steps to perform from ta to tb. Step size h is
            determined by
            
                    tb - ta
            h = ----------------
                        N
        
        Returns
        -------
        phase_solution : array_like
            Array of solutions in the specified space. The columns represent
            the state of the system at each time step for each variable.
        """
        f = self.f
        r0 = np.array(self.init_conds, float)
        h = (tb-ta)/N
        tpoints = np.arange(ta,tb,h)
        phase_solution = np.zeros(tpoints.shape + r0.shape, float)
        r = r0
        for i,t in enumerate(tpoints):
            phase_solution[i] = r
            s1 = h*f(r,t)
            s2 = h*f(r + 1/2*s1, t + 1/2*h)
            s3 = h*f(r + 1/2*s2, t + 1/2*h)
            s4 = h*f(r + s3, t + h)
            r += 1/6*(s1 + 2*s2 + 2*s3 + s4)
        self.h = h
        self.phase_solution = phase_solution
        self.t = tpoints

def system(t, l1, l2, k1, k2):
    x = k1*np.exp(l1*t) + k2*np.exp(l2*t)
    dx = k1*l1*np.exp(l1*t) + k2*l2*np.exp(l2*t)
    return x, dx

t = np.linspace(0, 10, 1000)
```

```python
x, dx = system(t, -1, -30, 31, 16)

fig = mp.figure()
ax = fig.add_subplot(111)
ax.plot(x, dx, "b.")
ax.set_xlabel("$x$")
ax.set_ylabel("$\dot{x}$")
ax.set_title("Attractor")

mp.savefig("attractor.png", dpi=300, bbox_inches="tight")
mp.show()
```

```python
x, dx = system(t, 1, 30, 31, 16)

fig = mp.figure()
ax = fig.add_subplot(111)
ax.plot(x, dx, "b.")
ax.set_xlabel("$x$")
ax.set_ylabel("$\dot{x}$")
ax.set_title("Repellor")

mp.savefig("repellor.png", dpi=300, bbox_inches="tight")
mp.show()
```

```python
x, dx = system(t, -3+1j, -1+6j, 31, 16)

fig = mp.figure()
ax = fig.add_subplot(111)
ax.plot(x.real, dx.real, "b.")
ax.set_xlabel("$x$")
ax.set_ylabel("$\dot{x}$")
ax.set_title("Stable focus")

mp.savefig("stable focus.png", dpi=300, bbox_inches="tight")
mp.show()
```

```python
x, dx = system(t, 1+1j, 3+6j, 31, 16)

fig = mp.figure()
ax = fig.add_subplot(111)
ax.plot(x.real, dx.real, "b.")
ax.set_xlabel("$x$")
ax.set_ylabel("$\dot{x}$")
ax.set_title("Unstable focus")

mp.savefig("unstablefocus.png", dpi=300, bbox_inches="tight")
mp.show()
```

```python
x, dx = system(t, 1j, 3j, 31, 16)

fig = mp.figure()
ax = fig.add_subplot(111)
ax.plot(x.real, dx.real, "b.")
ax.set_xlabel("$x$")
ax.set_ylabel("$\dot{x}$")
ax.set_title("Limit cycle")

mp.savefig("limit cycle.png", dpi=300, bbox_inches="tight")
mp.show()
```
