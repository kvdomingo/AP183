```python
import numpy as np
import matplotlib.pyplot as mp
import matplotlib.ticker as tick
from jupyterthemes import jtplot
```

```python
jtplot.reset()
mp.rcParams["font.family"] = "serif"
mp.rcParams["text.usetex"] = True
mp.rcParams["figure.figsize"] = (5*16/9, 5)
mp.rcParams["figure.dpi"] = 100
```

```python
w = np.logspace(-1, 6, 500)

class TransferFunc:
    
    def __init__(self, num, den, w):
        self.num = np.array(num)
        self.den = np.array(den)
        self.w = w
        self.numtf = 0
        self.dentf = 0
        s = 1j*w
        
        num = np.flip(self.num)
        idx = np.flip(range(len(num)))
        for i,a in enumerate(num):
            self.numtf += a * s**idx[i]
        den = np.flip(self.den)
        idx = np.flip(range(len(den)))
        for i,a in enumerate(self.den):
            self.dentf += a * s**idx[i]
            
        self.tf = self.numtf/self.dentf
        self.mag = np.abs(self.tf)
        self.phase = np.arctan2(self.tf.imag, self.tf.real)
        
    def bode(self, title=None, save=False):
        fig = mp.figure(figsize=(5*16/9, 5))
        
        ax = fig.add_subplot(212)
        ax.plot(self.w, np.degrees(self.phase), "k-")
        ax.set_xscale("log")
        ax.grid(True, which="both", linestyle="--")
        ax.set_xlabel("frequency, rad/s")
        ax.set_ylabel("phase, deg")
        
        ax = fig.add_subplot(211, sharex=ax)
        ax.plot(self.w, 20*np.log10(self.mag), "k-")
        ax.xaxis.set_major_formatter(tick.NullFormatter())
        ax.set_xscale("log")
        ax.grid(True, which="both", linestyle="--")
        ax.set_ylabel("magnitude, dB")
        
#         if title is not None:
#             ax.set_title(title)
        if save:
            mp.savefig(title + ".png", dpi=300, bbox_inches="tight")
        
        mp.show()
```

```python
numlist = np.array([[10], [10, 0], [5], [1], [10], [1, 2], [1], [5, 0], [2]])
denlist = np.array([[1, 0], [1], [1, 5], [1, 5], [1, 502, 100], [1, 500], [1, 5, 0], [1, 5], [1, 2, 100]])
assert(len(numlist) == len(denlist))
```

```python
for i in range(len(numlist)):
    G = TransferFunc(numlist[i], denlist[i], w)
    G.bode(save=True, title=str(i))
```
